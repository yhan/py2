```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate x values
arr = np.round(np.arange(-2, 2, 0.01), 2)
df = pd.DataFrame(arr, columns=['x'])

# Compute functions
df['y'] = df['x'] ** 2
df['z3'] = df['x'] ** 3

"""
# df['logx'] = np.where(df['x'] > 0, np.log(df['x']), np.nan)  # Avoid log of non-positive

   .venv\Lib\site-packages\pandas\core\arraylike.py:399: RuntimeWarning:
   divide by zero encountered in log


   looks correct, but the RuntimeWarning you're seeing is still triggered because of how np.where() works internally.
   np.where(condition, A, B) evaluates both A and B eagerly — even if the condition would exclude one of them.

   So in your case:

   np.where(df['x'] > 0, np.log(df['x']), np.nan)
   Still tries to evaluate np.log(df['x']) on the entire array, including x = 0 or x < 0, before applying the condition. That’s what causes the warning.

"""



"""
df.loc[df['x'] > 0, 'logx'] :
     in df select rows where x > 0 and label it as 'logx'
df.loc[df['x'] > 0, 'x']
     in df select rows where x > 0 and label it as 'x' ( x exists already in df as a column )

the assign ``=`
     will create logx column and the values are the np.log ...
"""
df['logx'] = np.nan
df.loc[df['x'] > 0, 'logx'] = np.log(df.loc[df['x'] > 0, 'x'])

df

# Plot
plt.figure(figsize=(12, 6))
plt.plot(df['x'], df['y'], label='x²', color='blue')
plt.plot(df['x'], df['z3'], label='x³', color='orange')
plt.plot(df['x'], df['logx'], label='log(x)', color='green')

plt.xlabel('x')
plt.ylabel('Value')
plt.title('x², x³, and log(x)')
plt.legend()
plt.grid(True)
plt.show()
```
